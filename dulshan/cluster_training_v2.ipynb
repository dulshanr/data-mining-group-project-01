```python
 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_parquet('./cleaned.parquet')
```

```python
num_columns = df.select_dtypes(include=['int', 'float']).columns
```

```python
scaler = preprocessing.StandardScaler()
```

```python
df[num_columns] = scaler.fit_transform(df[num_columns])
```

```python
df=df[num_columns]
```

```python
df_x = df.sample(n=5000)
```

```python
def cluster_agg(df, k, metric_affinity):
    cluster_model = AgglomerativeClustering(n_clusters=k,
                                            affinity=metric_affinity,
                                            linkage='average')
 
    clusters = cluster_model.fit_predict(df)
 
    score = silhouette_score(df,
                             cluster_model.labels_,
                             metric='euclidean')
    return clusters, score
```

```python
y_hd_euclidean, euclidean_score_hd = cluster_agg(df_x, 100, 'euclidean')
euclidean_score_hd
```
