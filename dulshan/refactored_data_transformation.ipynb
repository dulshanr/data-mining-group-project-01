```python
transformed_df = union_df.drop(columns=['maxdbddpdtollast6m_4187119P'], inplace=False).copy()
```

```python
cat_columns = transformed_df.select_dtypes(include=['object', 'bool'])
num_columns = transformed_df.select_dtypes(exclude=['object', 'bool'])
```

```python
num_columns_list = num_columns.columns.to_list()
cat_columns_list = cat_columns.columns.to_list()
```

```python
exclude_list = ['case_id','target', 'MONTH', 'WEEK_NUM']
exclude_num_list = list(filter(lambda i: i not in exclude_list, num_columns_list))
cat_exclude_list = ['datelastunpaid_3546854D','date_decision']
exclude_cat_list = list(filter(lambda i: i not in cat_exclude_list, cat_columns_list))
```

```python
exclude_cat_list
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()

df_scaled = std_scaler.fit_transform(transformed_df[exclude_num_list].to_numpy())
df_scaled = pd.DataFrame(df_scaled, columns=exclude_num_list)
```

```python
main_scaled = pd.concat([transformed_df[exclude_list], df_scaled], axis=1, join="inner")
```

```python
main_cat = transformed_df[exclude_cat_list].copy()
```

```python
from sklearn.feature_selection import SelectKBest 
from sklearn.feature_selection import chi2 
from sklearn.datasets import load_iris 
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
for col in main_cat.columns:
    main_cat[col] = label_encoder.fit_transform(main_cat[col])
```

```python
combined_df = pd.concat([main_scaled, main_cat], axis=1, join="inner")
```

```python
combined_df
```

```python
X = combined_df.drop(columns =['target'], inplace=False).to_numpy()
```

```python
all_columns_list = combined_df.drop(columns =['target'], inplace=False).columns.to_list()
```

```python
y= combined_df['target'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression


model = LogisticRegression(solver='lbfgs')
rfe = RFE(model, step = 20, n_features_to_select=20)
fit = rfe.fit(x_train, y_train)
print("Num Features: %d" % fit.n_features_)
print("Selected Features: %s" % fit.support_)
print("Feature Ranking: %s" % fit.ranking_)
```

```python
df_column_pairs = pd.DataFrame({'column_name': all_columns_list, 'condition': fit.support_})
```

```python
top_20_cols = df_column_pairs[df_column_pairs['condition']]['column_name'].to_list()
```
